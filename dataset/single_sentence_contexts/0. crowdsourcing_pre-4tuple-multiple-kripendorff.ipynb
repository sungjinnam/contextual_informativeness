```python
temp_scores = {}
for i in range(df_full_test.shape[0]):
#     print(i)
    sel = np.array(['s1', 's2', 's3', 's4'])
    sel_neg, sel_pos = df_full_test.iloc[i][['which_sentence_has_the_least_contextual_constraints_with_respect_to_the_blank', 
                                             'which_sentence_has_the_most_contextual_constraints_with_respect_to_the_blank']]
    sel_neu = np.array(sel)[~np.isin(sel, [sel_neg, sel_pos])]
    if sel_pos=='s1':
        if(df_full_test.iloc[i]['s1_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s1_idx']].append(1)
        else:
            temp_scores[df_full_test.iloc[i]['s1_idx']] = [1]
    if sel_pos=='s2':
        if(df_full_test.iloc[i]['s2_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s2_idx']].append(1)
        else:
            temp_scores[df_full_test.iloc[i]['s2_idx']] = [1]
    if sel_pos=='s3':
        if(df_full_test.iloc[i]['s3_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s3_idx']].append(1)
        else:
            temp_scores[df_full_test.iloc[i]['s3_idx']] = [1]
    if sel_pos=='s4':
        if(df_full_test.iloc[i]['s4_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s4_idx']].append(1)
        else:
            temp_scores[df_full_test.iloc[i]['s4_idx']] = [1]

            
    if sel_neg=='s1':
        if(df_full_test.iloc[i]['s1_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s1_idx']].append(-1)
        else:
            temp_scores[df_full_test.iloc[i]['s1_idx']] = [-1]
    if sel_neg=='s2':
        if(df_full_test.iloc[i]['s2_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s2_idx']].append(-1)
        else:
            temp_scores[df_full_test.iloc[i]['s2_idx']] = [-1]
    if sel_neg=='s3':
        if(df_full_test.iloc[i]['s3_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s3_idx']].append(-1)
        else:
            temp_scores[df_full_test.iloc[i]['s3_idx']] = [-1]
    if sel_neg=='s4':
        if(df_full_test.iloc[i]['s4_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s4_idx']].append(-1)
        else:
            temp_scores[df_full_test.iloc[i]['s4_idx']] = [-1]

    if sel_neu[0]=='s1':
        if(df_full_test.iloc[i]['s1_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s1_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s1_idx']] = [0]
    if sel_neu[0]=='s2':
        if(df_full_test.iloc[i]['s2_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s2_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s2_idx']] = [0]
    if sel_neu[0]=='s3':
        if(df_full_test.iloc[i]['s3_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s3_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s3_idx']] = [0]
    if sel_neu[0]=='s4':
        if(df_full_test.iloc[i]['s4_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s4_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s4_idx']] = [0]            

    if sel_neu[1]=='s1':
        if(df_full_test.iloc[i]['s1_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s1_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s1_idx']] = [0]
    if sel_neu[1]=='s2':
        if(df_full_test.iloc[i]['s2_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s2_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s2_idx']] = [0]
    if sel_neu[1]=='s3':
        if(df_full_test.iloc[i]['s3_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s3_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s3_idx']] = [0]
    if sel_neu[1]=='s4':
        if(df_full_test.iloc[i]['s4_idx'] in temp_scores):
            temp_scores[df_full_test.iloc[i]['s4_idx']].append(0)
        else:
            temp_scores[df_full_test.iloc[i]['s4_idx']] = [0]                        
```

```python
df_sents['scores'] = [temp_scores[i] if i in temp_scores else [-99] for i in df_sents['sent_idx']]

df_sents['scores_sum'] = [np.sum(scores) for scores in df_sents['scores']]
df_sents['scores_avg'] = [np.mean(scores) for scores in df_sents['scores']]
df_sents['scores_std'] = [np.std(scores) for scores in df_sents['scores']]

df_sents['scores_avg_rank'] = df_sents['scores_avg'].rank(method='average')

df_sents.to_csv("results_4tuple_multiple.csv")
```
